```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
CSV = 'Sales_Almaly5.csv'
HOST = 'https://krisha.kz/'
URL = 'https://krisha.kz/prodazha/kvartiry/almaty-almalinskij/?das[live.rooms]=5.100'
HEADERS = {'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
           'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.75 Safari/537.36'
    }
```

 Refers to the website page in order to obtain the content of the page (html)

```python
def get_html(url,params=''):
    r = requests.get(url, headers=HEADERS, params=params)
    return r
html = get_html(URL)
#print (html)
```

Take the content from the specified html

```python
def get_content(html):
    soup = BeautifulSoup(html,'html.parser')
    items = soup.find_all('div', class_='a-card__inc') # class 
    cards = []
    
    #print(items)
    for item in items:
        cards.append(
            {
                'title':item.find('div',class_='a-card__header-left').get_text(strip=True),
                'link_href': HOST + item.find('div',class_='a-card__header-left').find('a').get('href'),
                'price':item.find('div',class_='a-card__price').get_text(strip=True),
                'address':item.find('div',class_='a-card__wrapper-subtitle').get_text(strip=True),
                'date':item.find('div',class_='card-stats').get_text(strip=True),
                'comment':item.find('div',class_='a-card__text-preview').get_text(strip=True)
            
            }
        
        
        )
    return cards

```

Saves everything we paired in a file

```python
def save_doc(items,path):
        with open(path,'w',newline='', encoding = 'utf-8') as file:
            writer = csv.writer(file,delimiter =',')
            writer.writerow(['Заголовок','ссылка','Цена','Адрес','Дата','comment'])
            for item in items:
                writer.writerow([item['title'],item['link_href'],item['price'],item['address'],item['date'],item['comment']])
                
                
               
```

```python
def parser():
    PAGENATION = input('Укажите страницу для парсинга:  ')
    PAGENATION = int(PAGENATION.strip())
    html = get_html(URL)
    if html.status_code == 200:
        cards = []
        for page in range(1,PAGENATION):
            print(f'Парсим страницу: {page}')
            html = get_html(URL,params={'page':page})
            cards.extend(get_content(html.text))
            save_doc(cards,CSV)
        print(cards)
    else:
        print('Error')
```

```python
parser()
```
